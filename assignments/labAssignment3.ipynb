```python
import numpy as np
import matplotlib.pylab as plt
import librosa.display
import IPython.display
import librosa

#import audio files
!git clone https://github.com/jcdevaney/musicComputing2025.git
```

```python
def plotAudio2(sig,sampleRate,title):

    # add in the code from Assignment 2

    time = np.arange(0, sig.shape[0]) / sampleRate

    xlabel = 'Time (seconds)'
    ylabel = 'Amplitude'

    plt.plot(time, sig)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    plt.show()

```

```python
def plotAudioFreqDomain(sig,sr,title,winSize,specType):

    # create a normalized spectrogram
    # this includes several function calls
    # first create a basic specgrogram with librosa.stft() with the default window size (2048)
    # then make all of values in the spectrogram positive with np.abs()
    # then scale the spectogram to be in decibels, so it better reflects how we hear librosa.amplitude_to_db()

    # display the normalized spectrogram with a log frequency scale

    stft_result = librosa.stft(sig, n_fft=winSize)

    magnitude = np.abs(stft_result)

    db_spectrogram = librosa.amplitude_to_db(magnitude, ref=np.max)

    plt.figure(figsize=(6, 4))
    librosa.display.specshow(db_spectrogram, sr=sr, x_axis='time', y_axis=specType)
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"{specType.capitalize()} Spectrogram: {title}")
    plt.tight_layout()
    plt.show()
```

```python
def plotAudioFunc(sig,sr,title,winSize,specType):

   plotAudio2(sig,sr,title)

   plotAudioFreqDomain(sig, sr, title, winSize, specType)
```

```python
sig , sr = librosa.load('musicComputing2025/audioFiles/avm.wav')
```

```python
title = 'Ave Maria'
winSize = 2048
specType = 'linear'

plotAudioFunc(sig,sr,title,winSize,specType)


```

```python
title = 'Ave Maria'
winSize = 2048
specType = 'log'

plotAudioFunc(sig,sr,title,winSize,specType)
```
